```python
data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David'],
    'Age': [24, 27, 22, 32],
    'City': ['New York', 'Los Angeles', 'Chicago', 'Houston']
}

# Creating a DataFrame from the dictionary
dictonory_to_df= pd.DataFrame(data)
dictonory_to_df
data2 = {
    'Product': ['Laptop', 'Tablet', 'Smartphone', 'Headphones'],
    'Price': [1200, 300, 800, 150],
    'Stock': [30, 50, 100, 75]
}

# Creating a DataFrame from the second dictionary
dictonory_to_df2 = pd.DataFrame(data2)
dictonory_to_df2
```

```python
with pd.ExcelWriter('output_data.xlsx') as writer:
    dictonory_to_df.to_excel(writer, sheet_name='People', index=False)
    dictonory_to_df2.to_excel(writer, sheet_name='Products', index=False)
```
